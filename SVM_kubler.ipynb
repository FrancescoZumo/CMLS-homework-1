```python
import numpy as np
import librosa
import os
import matplotlib.pyplot as plt
import sklearn.svm as svm
from sklearn.svm import SVC
import IPython.display as ipd
import scipy as sp
%matplotlib inline
from sklearn.mixture import BayesianGaussianMixture
from sklearn.model_selection import train_test_split
import seaborn as sns
import pandas as pd
```

```python
# define the dictionary containing the 10 classes
class_train_files = {'zero': [], 'one': [], 'two': [], 'three': [], 'four': [], 'five': [], 'six': [], 'seven': [], 'eight': [], 'nine': []}

root = '../free-spoken-digit-dataset/recordings'
for f in os.listdir(root):
    for index, c in enumerate(class_train_files):
        if int(f[0]) == index:
            class_train_files[c].append(f)
```

```python
# listen to an audio file
audio, fs = librosa.load(os.path.join(root, class_train_files['eight'][2]), sr=None)
ipd.Audio(audio, rate=fs) # load the file
```

```python
df = pd.DataFrame(class_train_files)
# split the data into train and test set (for now 10% test)
train, test = train_test_split(df, test_size=0.1, random_state=42, shuffle=True)
print(train.shape[0])
print(test.shape[0])

x = np.empty(train.shape[0], object)
```

```python
# initialize dictionaries
dict_train_mfcc = {'zero': [], 'one': [], 'two': [], 'three': [], 'four': [], 'five': [], 'six': [], 'seven': [], 'eight': [], 'nine': []}
dict_test_mfcc = {'zero': [], 'one': [], 'two': [], 'three': [], 'four': [], 'five': [], 'six': [], 'seven': [], 'eight': [], 'nine': []}


# why 13? try different values. check warning
n_mfcc = 10

df_list = [train, test]

# for train and test
for index, data in enumerate(df_list):
    # for each class
    for c in data:
        # set up array for storing mfcc
        if index == 0:
            dict_train_mfcc[c] = np.zeros(((len(data[c])), n_mfcc))
        else:
            dict_test_mfcc[c] = np.zeros(((len(data[c])), n_mfcc))
        #temp_features = np.zeros(((len(data[c])), n_mfcc))
        # compute mfcc
        for index2, f in enumerate(data[c]):
            audio, fs = librosa.load(os.path.join(root, f), sr=None)
            mfcc = librosa.feature.mfcc(y=audio, sr=fs, n_mfcc=n_mfcc)
            #mfcc = compute_mfcc(audio, fs, n_mfcc)
            #temp_features[index2, :] = np.mean(mfcc, axis=1)
            # store results for current class and file
            if index == 0:
                dict_train_mfcc[c][index2, :] = np.mean(mfcc, axis=1)
            else:
                dict_test_mfcc[c][index2, :] = np.mean(mfcc, axis=1)
print('finished')
```

```python
x_dict_train = dict_train_mfcc['zero'].shape[0]
y_train = [np.zeros(x_dict_train)]
for i in np.arange(1,10):
    y_train.append(np.ones(x_dict_train)*i)

x_dict_test = dict_test_mfcc['zero'].shape[0]
y_test = [np.zeros(x_dict_test)]
for i in np.arange(1,10):
    y_test.append(np.ones(x_dict_test)*i)
    
y_train_conc = []

feat_max = np.max(np.concatenate((dict_train_mfcc['zero'], dict_train_mfcc['one'], dict_train_mfcc['two'],dict_train_mfcc['three'], dict_train_mfcc['four'], dict_train_mfcc['five'], dict_train_mfcc['six'], dict_train_mfcc['seven'], dict_train_mfcc['eight'], dict_train_mfcc['nine']), axis=0), axis=0)
feat_min = np.min(np.concatenate((dict_train_mfcc['zero'], dict_train_mfcc['one'], dict_train_mfcc['two'],dict_train_mfcc['three'], dict_train_mfcc['four'], dict_train_mfcc['five'], dict_train_mfcc['six'], dict_train_mfcc['seven'], dict_train_mfcc['eight'], dict_train_mfcc['nine']), axis=0), axis=0)

```

```python
dict_train_mfcc_norm = {'zero': [], 'one': [], 'two': [], 'three': [], 'four': [], 'five': [], 'six': [], 'seven': [], 'eight': [], 'nine': []}
dict_test_mfcc_norm = {'zero': [], 'one': [], 'two': [], 'three': [], 'four': [], 'five': [], 'six': [], 'seven': [], 'eight': [], 'nine': []}

for key, val in dict_train_mfcc.items():
    for a in val:
        dict_train_mfcc_norm[key].append( (a - feat_min) / (feat_max - feat_min) )
        
for key, val in dict_test_mfcc.items():
    for a in val:
        dict_test_mfcc_norm[key].append( (a - feat_min) / (feat_max - feat_min) )

```

```python
SVM_parameters={
    'C': 1,
    'kernel': 'rbf',
}

```

```python
clf = []
ten = ['zero','one','two','three','four','five','six','seven','eight','nine']
for i in np.arange(10):
    for j in np.arange(10):
        if (j > i):
            clf_svm = svm.SVC(**SVM_parameters, probability=True)
            clf.append(clf_svm.fit(np.concatenate((dict_train_mfcc_norm[ten[i]],dict_train_mfcc_norm[ten[j]]),axis=0), np.concatenate((y_train[i],y_train[j]),axis=0)))
        else:
            clf.append(0)
```
